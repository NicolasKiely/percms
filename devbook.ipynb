```python
import crypto.models
import crypto.utils

# Create backtest config
base_name = 'BTC'
exchange_name = 'Poloniex'
pair_names = ['ETH']
start_date = '2017-10-01 00:00:00'
end_date = '2017-12-30 00:00:00'
period = 14400

exchange = crypto.models.Exchange.objects.get(name=exchange_name)
base_currency = crypto.models.Currency.objects.get(symbol=base_name)
pairs = [
    crypto.utils.get_currency_pair(exchange_name, base_name, p)
    for p in pair_names
]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Load data
import pandas as pd

df_list = []

for pair in pairs:
    name=pair.c2
    candles = pair.candle_stick_set.filter(
        period=period,
        stamp__gte=start_date,
        stamp__lt=end_date
    ).order_by('stamp').all()[:]
    
    if len(candles) == 0:
        continue
        
    candles_df = pd.DataFrame({
        name+'_open': [c.p_open for c in candles],
        name+'_close': [c.p_close for c in candles],
        name+'_high': [c.p_high for c in candles],
        name+'_low': [c.p_low for c in candles],
        name+'_volume': [c.volume for c in candles],
    }, index = [c.stamp for c in candles])
    df_list.append(candles_df)
    
df = pd.concat(df_list, axis=1)
print df.head()
```

```python
def highest(series, t_length):
    result = series.iloc[-1]
    data_len = len(series)
    if t_length >= data_len:
        subset = series
    else:
        subset = series.iloc[data_len-t_length:]
        
    for x in subset:
        result = max(x, result)
    return result

def lowest(series, t_length):
    result = series.iloc[-1]
    data_len = len(series)
    if t_length >= data_len:
        subset = series
    else:
        subset = series.iloc[data_len-t_length:]
        
    for x in subset:
        result = min(x, result)
    return result

def latest(ds, t_length):
    data_len = len(ds)
    return ds if t_length >= data_len else ds.iloc[data_len-t_length:]

def ema(source, t_length):
    alpha = 2.0 / (t_length + 1)
    prev_val = source[0]
    results = []
    for x in source:
        new_val = alpha*x + (1-alpha)*prev_val
        results.append(new_val)
        prev_val = new_val
    return results

def cross(xs, ys):
    for i in range(0, len(xs)-1):
        if (xs[i] <= ys[i]) and (xs[i+1] > ys[i+1]):
            return True
        elif (xs[i] >= ys[i]) and (xs[i+1] < ys[i+1]):
            return True
    return False


def band_strat(data, do_plot):
    # Candle sticks to look behind
    history_range = 6 * 60
    hi_lo_length = 4
    radius = 8
    
    p_close = latest(data['close'], history_range).tolist()
    
    # Calculate hi/lo bands
    lo_band = []
    hi_band = []
    lo_ema = []
    hi_ema = []
    norm_close = []
    
    last_lo_ema = p_close[0]
    last_hi_ema = p_close[0]
    for i, c in enumerate(p_close):
        # HI/LOW
        high_val = c
        low_val = c
        back = i - hi_lo_length
        if back < 0:
            back = 0
        
        for j in range(back, i):
            k = p_close[j]
            if k < low_val:
                low_val = k
            if k > high_val:
                high_val = k
        
        # EMA
        if high_val == low_val:
            norm_c = 0.5
        else:
            norm_c = (c - low_val) / (high_val - low_val)
        
        norm_close.append(norm_c)
        lo_alpha = 0.5*(1-norm_c)
        hi_alpha = 0.5*(norm_c)
        new_lo_ema = last_lo_ema * (1-lo_alpha) + lo_alpha * c
        new_hi_ema = last_hi_ema * (1-hi_alpha) + hi_alpha * c
        
        last_lo_ema = new_lo_ema
        last_hi_ema = new_hi_ema
        
        lo_band.append(low_val)
        hi_band.append(high_val)
        lo_ema.append(last_lo_ema)
        hi_ema.append(last_hi_ema)
        
    
    if do_plot:
        # Band graphs
        plt.figure(1)
        plt.title('Close')
        plt.plot(data['close'], color='black')
        plt.plot(lo_band, color='blue')
        plt.plot(hi_band, color='blue')
        plt.plot(lo_ema, color='red')
        plt.plot(hi_ema, color='red')
        
        # Buy/Sell graphs
        #plt.rcParams["figure.figsize"] = (12, 9)
        #plt.figure(2)
        #plt.plot(data['close'], color='black')
    
    last_price = p_close[-1]
    if len(p_close) <= 2:
        return {'buy': False, 'sell': False}
    
    buy, sell = False, False
    #if  (lo_ema[-1] > 0.95 * lo_ema[-2]):
    #    if (hi_ema[-2] >= hi_band[-2]) and (hi_ema[-1] < hi_band[-1]):
    #        buy = True
    if (p_close[-3] + p_close[-2] <= 1.01 * (lo_ema[-3] + lo_ema[-2])):
        if (p_close[-1] >= p_close[-2]):
            buy = True
    
    if (hi_ema[-2] < hi_band[-2]) and (hi_ema[-1] >= hi_band[-1]):
        sell = True
        #buy = False
        
    return {'buy': buy, 'sell': sell}


def test_to_ith_row(df, c_name, i, strategy, do_plot=False):
    trunc_df = df.iloc[:i+1]
    data = pd.DataFrame({
        'open': trunc_df[c_name+'_open'],
        'close': trunc_df[c_name+'_close'],
        'high': trunc_df[c_name+'_high'],
        'low': trunc_df[c_name+'_low'],
        'volume': trunc_df[c_name+'_volume']
    }, index = trunc_df.index)
    
    # Strategy
    return strategy(data, do_plot)

results = []
for i in range(1, len(df)):
    do_plot = ( i + 1 == len(df) )
    results.append(test_to_ith_row(df, 'ETH', i, band_strat, do_plot))
```
